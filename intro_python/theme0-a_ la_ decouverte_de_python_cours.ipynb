## Les variables

Les types simples :
- Booléen : **bool** ->  Vrai ou faux (True / False)
- Caractère : **str**
- Entier : **int**
- Réel : **float**

````python
A = 3
B = A + 3
A = 59
# à la fin de ce programme, quelle est la valeur de A ? de B ?
print(A)
print(B)
````

## Les tests et conditions 
Les conditions sont un concept essentiel en programmation.
Dans l’instruction si condition alors bloc, la condition est une expression booléenne, et le bloc n’est exécuté que si la condition est vraie.

Exemples :

````python
a = 5
if a > 0: 
      print("a est supérieur à 0.")
        
age = 21
if age >= 18: 
    print("Vous êtes majeur.")
elif age < 18:
    print("Vous êtes mineur.")
else: 
    print("Erreur")
````

### la boucle for
L’instruction for est utilisée lorsque le nombre d’itérations est connu à l’avance: elle initialise un compteur, l’incrémente après chaque exécution du bloc d’instructions, et vérifie que le compteur ne dépasse pas la borne supérieure.

Exemple :

````python
n = int(input("Saisir n : "))
s = 0
for i in range(1, n):
    s = s + i
print(s)

# Que fait cet algorithme ?
````

!!! info Information sur range()

!!!

### La boucle while
La boucle while est utilisée lorsque le nombre d’itérations n’est pas connu à l’avance: elle exécute le bloc d’instructions tant que la condition reste vraie.
Exemple:

````python
n = 1
s = 0
while n != 0:
    n = int(input("Saisir n : "))
    s = s + n 
print(s)

# Que fait cet algorithme ?
````

## Les fonctions
Une fonction est un bloc de code qui a un nom et que l’on peut réutiliser plusieurs fois.
Au lieu de répéter les mêmes instructions, on les met dans une fonction et on “appelle” cette fonction quand on en a besoin.

!!! example Exemple
``` python
def dire_bonjour():
    print("Bonjour !")
```
!!!

````python
#Exemple simple
def dire_bonjour():
    print("Bonjour !")

# Appel de la fonction
dire_bonjour()


#La table de 7
def table_par_7():
    nb = 7
    i = 0 
    while i < 10: # Tant que i est strictement inférieure à 10,
        print(i + 1, "*", nb, "=", (i + 1) * nb)
        i += 1 # On incrémente i de 1 à chaque tour de boucle.
        
# Appel de la fonction
table_par_7()
````

### Les paramètres et arguments des fonctions

Une fonction peut recevoir des informations extérieures (des données) pour travailler avec.

- Paramètres : ce sont les noms utilisés dans la définition de la fonction.
- Arguments : ce sont les valeurs concrètes qu’on donne à la fonction quand on l’appelle.

````python
# Exemple simple 
def dire_bonjour_a(nom):   # "nom" est un paramètre
    print("Bonjour", nom)

# Ici "Eren" est un argument lors de l'appel
dire_bonjour_a("Eren")

# Ici "Mikasa" est un argument lors de l'appel
dire_bonjour_a("Mikasa")


#La table de multiplication
def table_multiplication(table):
    i = 0 
    while i < 10: # Tant que i est strictement inférieure à 10,
        print(i + 1, "*", table, "=", (i + 1) * table)
        i += 1 # On incrémente i de 1 à chaque tour de boucle.
        
# Appel de la fonction pour la table de 2
table_multiplication(2)
# Appel de la fonction pour la table de 8
table_multiplication(8)
````
